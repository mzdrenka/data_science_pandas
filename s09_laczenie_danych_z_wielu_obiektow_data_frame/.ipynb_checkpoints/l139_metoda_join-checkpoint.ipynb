```python
import pandas as pd
import numpy as np
```

```python
customer_names = pd.read_csv('../zasoby/nw/customers.csv', usecols=['CustomerID', 'CompanyName'])
customer_cities = pd.read_csv('../zasoby/nw/customers.csv', usecols=['CustomerID', 'City', 'Country'])
```

```python
len(customer_names)
```

```python
len(customer_cities)
```

```python
customer_names.set_index(keys='CustomerID', inplace=True)
customer_cities.set_index(keys='CustomerID', inplace=True)
```

```python
customer_names.join(customer_cities)
```

```python
customer_names.reset_index()
```

```python
customer_cities.join(other=customer_names, on="CustomerID")
```
