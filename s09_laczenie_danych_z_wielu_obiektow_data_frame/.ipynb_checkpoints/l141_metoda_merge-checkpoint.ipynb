```python
import pandas as pd
import numpy as np
```

```python
f_2015 = pd.read_csv('../zasoby/Fortune_500_2015.csv', usecols=['rank', 'company'])
f_2016 = pd.read_csv('../zasoby/Fortune_500_2016.csv', usecols=['rank', 'company'])
```

```python
companies_2015_2016 = f_2015.merge(f_2016, on='company', how='inner')
```

```python
companies_2015_2016.head()
```

```python
companies_2015_2016 = f_2015.merge(f_2016, on='company', how='inner', suffixes=['_2015', '_2016'])
```

```python
companies_2015_2016.head()
```

```python
companies_2015_2016.loc[companies_2015_2016["company"] == "Microsoft"]
```

```python
companies_2015_2016 = f_2015.merge(f_2016, on='company', how='outer', suffixes=['_2015', '_2016'], indicator=True)
```

```python
companies_2015_2016.head(100)
```

```python
only_2015 = companies_2015_2016.loc[companies_2015_2016["_merge"] == "left_only"]
```

```python
only_2015
```

```python
only_2016 = companies_2015_2016.loc[companies_2015_2016["_merge"] == "right_only"]
only_2016.head()
```
