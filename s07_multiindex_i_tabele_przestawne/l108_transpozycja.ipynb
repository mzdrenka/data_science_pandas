```python
import pandas as pd
import numpy as np
```

```python
fuel = pd.read_csv('../zasoby/fuel.csv', usecols=['Year','Make','Model','Range (FT1)','City Range (FT1)'], index_col=['Year','Make','Model'])
```

```python
fuel.sort_index(inplace=True)
```

```python
fuel.head()
```

```python
fuel.loc[1984]
```

```python
fuel.loc[1984, "Alfa Romeo"]
```

```python
fuel.loc[1984,'Alfa Romeo','Spider Veloce 2000']
```

```python
fuel.loc[(1984,'Alfa Romeo','Spider Veloce 2000'), 'City Range (FT1)']
```

```python
fuel = fuel.transpose()
```

```python
fuel.head()
```

```python
fuel.loc['Range (FT1)']
```

```python
fuel.loc['Range (FT1)', (1984)]
```
