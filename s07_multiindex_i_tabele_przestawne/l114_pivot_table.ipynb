```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('../zasoby/marathon_results_2016.csv', usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'], index_col='Bib')
```

```python
df.head()
```

```python
df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)
 
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
 
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
 
df['TotalSeconds'] = df['TotalSeconds'].replace(0,np.nan)
df['HalfSeconds'] = df['HalfSeconds'].replace(0,np.nan)
```

```python
df.head()
```

```python
df.pivot_table(values='TotalSeconds', index='Age', columns='M/F', aggfunc='mean')
```

```python
df.pivot_table(values=['HalfSeconds','TotalSeconds'], index='Age', columns='M/F', aggfunc='mean')
```

```python
piv_paf = df.pivot_table(values=['HalfSeconds','TotalSeconds'], index='Age', columns='M/F', aggfunc='mean')
piv_paf.swaplevel(axis='columns').sort_index(axis=1)
```

```python
df.pivot_table(index='Age', values='HalfSeconds', columns='M/F', aggfunc='count')
```

```python
pif_paf_c = df.pivot_table(index='Age', values='HalfSeconds',  aggfunc='count')
```

```python
pif_paf_c.rename(columns={"HalfSeconds": "Count"})
```
