```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline

marathon = pd.read_csv("../zasoby/marathon_results_2017.csv", usecols=["Age","M/F","Country","40K"])
marathon["TimeSeconds"] = marathon["40K"].apply(lambda x: pd.to_timedelta(x, errors='coerce').total_seconds())
marathon.head(5)
```

```python
marathon.plot(kind="scatter", x="Age", y="TimeSeconds")
```

```python
is_man = marathon["M/F"] == "M"
marathon_m = marathon[is_man]
marathon_m.head(5)
```

```python
is_woman = marathon["M/F"] == "F"
marathon_w = marathon[is_woman]
marathon_w.head(5)
```

```python
ax = marathon_m.plot.scatter(x='Age', y='TimeSeconds', color='Blue', label='Men')
marathon_w.plot.scatter(x='Age', y='TimeSeconds', color='Pink', label='Women', ax=ax)
```

```python
marathon_m.plot.hexbin(x="Age",y="TimeSeconds",gridsize=30)
```

```python
marathon_w.plot.hexbin(x="Age",y="TimeSeconds",gridsize=10)
```
