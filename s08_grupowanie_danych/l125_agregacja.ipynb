```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('../zasoby/marathon_results_2016.csv', index_col='Bib', usecols=['Bib','40K','Half','Pace','Age','M/F','Country', 'State','City'])
df = df[(df['40K'] != '-') & (df['Half'] != '-')]
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
group_city = df.groupby(by="City")
```

```python
group_city.mean(numeric_only=True).head()
```

```python
group_city[["TotalSeconds","HalfSeconds"]].mean(numeric_only=True).head()
```

```python
group_city["TotalSeconds"].sum(numeric_only=True).head()
```

```python
group_age = df.groupby(by="Age")
```

```python
group_age.count().head()
```

```python
group_age['TotalSeconds'].count().sort_values(ascending=False)
```
