```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('../zasoby/marathon_results_2016.csv', index_col='Bib', usecols=['Bib','40K','Half','Pace','Age','M/F','Country', 'State','City'])
df = df[(df['40K'] != '-') & (df['Half'] != '-')]
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

Podobnie jak w materiale video, przygotuj pętlę, która dla każdego elementu z listy cities pobierze z oryginalnego obiektu df wiersze pasujące do w danej chwili przetwarzanego miasta. Do zmiennej groups dodaj nową parę klucz/wartość, gdzie kluczem jest nazwa kraju, a wartością wszystkie te wiersze z obiektu df, które opisują wyniki maratonów w tych miastach.

```python
groups = {}
```

```python
for city in cities:
    tmp_df = df.loc[ df["City"] == city]
    groups[city] = tmp_df
```

```python
groups
```

```python
groups["San Francisco"].describe()
```
