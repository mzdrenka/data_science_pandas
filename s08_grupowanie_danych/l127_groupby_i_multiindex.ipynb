```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('../zasoby/marathon_results_2016.csv', index_col='Bib', usecols=['Bib','40K','Half','Pace','Age','M/F','Country', 'State','City'])
df = df[(df['40K'] != '-') & (df['Half'] != '-')]
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
country_city = df.groupby(by=["Country","City"])
```

```python
country_city.size()
```

```python
df[df["Country"] == "POL"]["City"].value_counts()
```

```python
country_city.mean(numeric_only=True)
```

```python
country_city.get_group(('USA','San Francisco')).head()
```

```python
sex_age = df.groupby(by=["M/F","Age"])
```

```python
sex_age.head()
```

```python
sex_age.size()
```

```python
sex_age.mean(numeric_only=True)
```

```python
sex_age.get_group(("M", 25)).head()
```
