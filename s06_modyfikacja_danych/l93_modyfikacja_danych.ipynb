```python
import pandas as pd
```

```python
professions = pd.read_csv('../zasoby/Prestige.csv', index_col='name')
professions.head()
```

```python
professions.loc['chemists', 'type'] = 'scientsts'
professions.head()
```

```python
isScientist = professions['type'] == 'scientsts'
```

```python
isScientist
```

```python
professions.loc[isScientist, 'prestige'] = 90
```

```python
professions.head()
```

```python
professions['income'] = professions['income'].astype('float32')
professions.loc[isScientist, 'income'] = professions['income'] * 1.5
```
