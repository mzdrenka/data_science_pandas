```python
import pandas as pd
import numpy as np
```

```python
animals = pd.read_csv('../zasoby/sleep_time.csv', index_col='ID').fillna(0)
```

```python
animals.head()
```

```python
herbi = animals.where(animals['vore']=='herbi')
```

```python
len(herbi)
```

```python
herbi.dropna(inplace=True)
```

```python
len(herbi)
```

```python
cerni_animals = animals['vore'] == 'carni'
carni = animals[cerni_animals]
```

```python
len(carni)
```

```python
omni_animals = animals['vore'] == 'omni'
omni = animals[omni_animals]
len(omni)
```

```python
animals['vore'].value_counts()
```

```python
print('Sredni czas snu herbi: {} '.format(herbi['sleep_total'].mean()))
print('Sredni czas snu omni: {} '.format(omni['sleep_total'].mean()))
print('Sredni czas snu carni: {} '.format(carni['sleep_total'].mean()))
```

```python
animals.query('sleep_total > 14')
```
