```python
import pandas as pd
import numpy as np
```

```python
fuel = pd.read_csv('../zasoby/fuel.csv', low_memory=False, usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)'], index_col='Vehicle ID')
fuel.head()
```

```python
fuel.loc[27705, 'Class'] = np.NaN
fuel.loc[26561, 'Class'] = np.NaN
fuel.loc[27550, 'Fuel Type'] = np.NaN
fuel.loc[27705, 'Combined MPG (FT1)'] = np.NaN
fuel.loc[27681, 'Combined MPG (FT1)'] = np.NaN
```

```python
fuel.head()
```

```python
fuel.fillna(value=-1).head()
```

```python
replace_rules = {"Class": "---", 
                 "Fuel Type": "---",
                 "Combined MPG (FT1)": "-1"}
```

```python
fuel.fillna(value=replace_rules).head()
```

```python
avg_MPG = round(fuel['Combined MPG (FT1)'].mean(), 1)
```

```python
avg_MPG
```

```python
fuel.head()
```

```python
fuel['Class'].fillna(value='?', inplace=True)
fuel['Fuel Type'].fillna(value='?', inplace=True)
fuel['Combined MPG (FT1)'].fillna(value=avg_MPG, inplace=True)
fuel.head()
```

```python
fuel.loc[27705, 'Class'] = np.NaN
fuel.loc[26561, 'Class'] = np.NaN
fuel.loc[27550, 'Fuel Type'] = np.NaN
fuel.loc[27705, 'Combined MPG (FT1)'] = np.NaN
fuel.loc[27681, 'Combined MPG (FT1)'] = np.NaN
fuel.head()
```
