```python
import pandas as pd
import numpy as np
```

```python
fortune = pd.read_csv('../zasoby/Fortune_500_2017.csv', usecols=['Rank','Title','Employees','Profits','Assets'], index_col='Rank')
fortune.head()
```

```python
fortune['RankByEmployee'] = fortune['Employees'].rank(ascending=False)
```

```python
fortune.head()
```

```python
fortune['RankByProfits'] = fortune['Profits'].rank(ascending=False)
```

```python
fortune.head()
```

```python
fortune.nlargest(columns=['Assets'], n=5)
```

```python
fortune.nsmallest(columns=['Assets'], n=5)
```
