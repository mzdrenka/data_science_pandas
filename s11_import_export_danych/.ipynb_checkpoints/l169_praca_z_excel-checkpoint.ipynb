```python
import pandas as pd
import numpy as np
```

```python
mcdonald = pd.read_csv('../zasoby/Mcdonalds.csv', usecols=['Category', 'Item', 'Calories', 'TotalFat'])
```

```python
with pd.ExcelWriter(path='../zasoby/mcdonalds-excel.xlsx', mode='w') as excelWriter:
     mcdonald.to_excel(excelWriter, sheet_name="Menu",index=False)
```

```python
imported1 = pd.read_excel('../zasoby/mcdonalds-excel.xlsx', sheet_name='Menu')
```

```python
imported1.head()
```

```python
imported1['Calories'] = imported1['Calories'] / 2
```

```python
imported1.head()
```

```python
with pd.ExcelWriter(path='../zasoby/mcdonalds-excel.xlsx', mode='a', engine="openpyxl") as excelWriter:
    imported1.to_excel(excelWriter, sheet_name='Light', index=False)
```

```python
imported1['Calories'] = imported1['Calories'] / 2
```

```python
with pd.ExcelWriter(path='../zasoby/mcdonalds-excel.xlsx', mode='a', engine='openpyxl') as excelWriter:
    imported1.to_excel(excelWriter, sheet_name='ExtraLight', index=False)
```
